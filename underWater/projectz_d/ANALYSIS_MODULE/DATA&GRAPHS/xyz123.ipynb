```python
import requests
import pandas as pd
import plotly.express as px

# 1. Setup the Open-Meteo API request
url = "https://api.open-meteo.com/v1/forecast"
params = {
    "latitude": 28.5616,  # Set to local NCR coordinates
    "longitude": 77.2802,
    "hourly": ["temperature_2m", "relative_humidity_2m"],
    "timezone": "auto",
    "forecast_days": 3
}

# 2. Fetch the data from the API
print("Fetching weather data...")
response = requests.get(url, params=params)
data = response.json()

# 3. Process the JSON response into a Pandas DataFrame
# DataFrames make data manipulation much easier than raw lists or dictionaries
hourly_data = data["hourly"]
df = pd.DataFrame({
    "Time": pd.to_datetime(hourly_data["time"]),
    "Temperature (°C)": hourly_data["temperature_2m"],
    "Humidity (%)": hourly_data["relative_humidity_2m"]
})

print("Data loaded! Generating chart...")

# 4. Create an interactive line chart using Plotly Express
fig = px.line(
    df, 
    x="Time", 
    y=["Temperature (°C)", "Humidity (%)"], # Plotting two lines at once
    title="3-Day Hourly Forecast",
    markers=True,                 # Adds dots to the data points
    template="plotly_dark",       # Looks great in dark-themed notebooks (use "plotly_white" for light)
    labels={"value": "Metrics", "variable": "Legend"} 
)

# Make the hover data look a bit cleaner
fig.update_traces(hovertemplate='%{y}')
fig.update_layout(hovermode="x unified") # Shows all values for a specific time when you hover

# 5. Render the interactive plot
fig.show()
```
