```python
import json
import base64
import numpy as np
import os

def export_seismic_dashboard_html(filename='seismic_dashboard_interactive.html', output_dir=None):
    """Export as standalone HTML with 50-50 layout: map + trend graph + 3D image"""
    
    # Convert to list of dicts for proper serialization
    data_records = df_100.to_dict('records')
    
    # Ensure all numeric values are proper JSON-serializable types
    for record in data_records:
        for key, value in record.items():
            if isinstance(value, np.integer):
                record[key] = int(value)
            elif isinstance(value, np.floating):
                record[key] = float(value)
            elif isinstance(value, np.ndarray):
                record[key] = value.tolist()
    
    # Prepare images
    image_assets = {}
    image_files = {
        'delhi_2007': 'delhi 2007.png',
        'kutch_2007': 'kutch 2007.png'
    }
    
    for asset_name, file_path in image_files.items():
        if os.path.exists(file_path):
            try:
                with open(file_path, 'rb') as img_file:
                    b64_data = base64.b64encode(img_file.read()).decode()
                    image_assets[asset_name] = f"data:image/png;base64,{b64_data}"
                    print(f"✅ Embedded image: {asset_name}")
            except Exception as e:
                print(f"❌ Failed to read {file_path}: {e}")
                image_assets[asset_name] = None
        else:
            image_assets[asset_name] = None
    
    # Create HTML with 50-50 layout
    data_json = json.dumps(data_records, indent=2)
    terrain_json = json.dumps(TERRAIN_PROFILES)
    images_json = json.dumps(image_assets)
    
    html_content = f"""<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Seismic Dashboard - Interactive</title>
    <script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
    <style>
        * {{
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }}
        
        html, body {{
            width: 100%;
            height: 100%;
        }}
        
        body {{
            background-color: #050505;
            color: #00d4ff;
            font-family: monospace;
            display: flex;
            flex-direction: column;
            overflow: hidden;
        }}
        
        .main-container {{
            display: flex;
            flex: 1;
            min-height: 0;
            gap: 5px;
            padding: 5px;
        }}
        
        .map-section {{
            flex: 0 0 50%;
            background-color: #050505;
        }}
        
        #seismic-map {{
            width: 100% !important;
            height: 100% !important;
        }}
        
        .analysis-section {{
            flex: 0 0 50%;
            display: flex;
            flex-direction: column;
            gap: 5px;
            background-color: #050505;
        }}
        
        .trend-panel {{
            flex: 0 0 50%;
            background-color: #0a0a0a;
            border: 1px solid #333;
            border-radius: 3px;
            min-height: 0;
        }}
        
        #seismic-trend-graph {{
            width: 100% !important;
            height: 100% !important;
        }}
        
        .image-panel {{
            flex: 0 0 50%;
            background-color: #0a0a0a;
            border: 1px solid #333;
            border-radius: 3px;
            display: flex;
            align-items: center;
            justify-content: center;
            min-height: 0;
        }}
        
        .image-panel img {{
            max-width: 100%;
            max-height: 100%;
            padding: 5px;
        }}
        
        .control-panel {{
            height: 60px;
            background-color: #0a0a0a;
            border-top: 2px solid #00d4ff;
            padding: 10px 50px;
            display: flex;
            align-items: center;
            gap: 15px;
            flex-shrink: 0;
        }}
        
        .slider-input {{
            flex: 1;
            max-width: 300px;
            height: 6px;
            background: #333;
            border: none;
            border-radius: 3px;
            outline: none;
            -webkit-appearance: none;
        }}
        
        .slider-input::-webkit-slider-thumb {{
            -webkit-appearance: none;
            appearance: none;
            width: 16px;
            height: 16px;
            border-radius: 50%;
            background: #00d4ff;
            cursor: pointer;
        }}
        
        .slider-input::-moz-range-thumb {{
            width: 16px;
            height: 16px;
            border-radius: 50%;
            background: #00d4ff;
            cursor: pointer;
            border: none;
        }}
        
        .year-label {{
            color: #00d4ff;
            font-weight: bold;
            font-size: 12px;
            min-width: 60px;
        }}
    </style>
</head>
<body>
    <div class="main-container">
        <div class="map-section">
            <div id="seismic-map"></div>
        </div>
        <div class="analysis-section">
            <div class="trend-panel">
                <div id="seismic-trend-graph"></div>
            </div>
            <div class="image-panel">
                <div id="geological-3d-slab" style="width: 100%; height: 100%; display: flex; align-items: center; justify-content: center;"></div>
            </div>
        </div>
    </div>
    
    <div class="control-panel">
        <label class="year-label">Year Range:</label>
        <input type="range" id="year-min" class="slider-input" min="2007" max="2017" value="2007">
        <span class="year-label" id="year-display">2007 - 2017</span>
        <input type="range" id="year-max" class="slider-input" min="2007" max="2017" value="2017">
    </div>

    <script>
        const earthquakeData = {data_json};
        const imageAssets = {images_json};
        
        console.log('📊 Data loaded:', earthquakeData.length, 'records');
        
        function getZoneName(city) {{
            const parts = city.split(' ');
            return parts.slice(0, -1).join(' ');
        }}
        
        function getImageForZone(zoneName) {{
            const zoneUpper = zoneName.toUpperCase();
            let imageKey = 'kutch_2007';
            if (zoneUpper.includes('DELHI')) {{
                imageKey = 'delhi_2007';
            }} else if (zoneUpper.includes('KUTCH')) {{
                imageKey = 'kutch_2007';
            }}
            return imageAssets[imageKey] || imageAssets['kutch_2007'];
        }}
        
        function displayGeologicalImage(cityName) {{
            const zoneName = getZoneName(cityName);
            const imagePanel = document.getElementById('geological-3d-slab');
            imagePanel.innerHTML = '';
            const imageDataUri = getImageForZone(zoneName);
            if (imageDataUri && imageDataUri.startsWith('data:')) {{
                const img = document.createElement('img');
                img.src = imageDataUri;
                img.style.maxWidth = '100%';
                img.style.maxHeight = '100%';
                imagePanel.appendChild(img);
            }} else {{
                imagePanel.innerHTML = '<div style="color: #999; text-align: center; font-size: 12px;">No image</div>';
            }}
        }}
        
        function filterByYears(minYear, maxYear) {{
            return earthquakeData.filter(d => d.Year >= minYear && d.Year <= maxYear);
        }}
        
        function generateMap(minYear, maxYear) {{
            const filteredData = filterByYears(minYear, maxYear);
            if (filteredData.length === 0) {{
                filteredData.push(earthquakeData[0]);
            }}
            
            const impactSizes = filteredData.map(d => Math.pow(d.Magnitude, 4));
            const maxSize = Math.max(...impactSizes);
            const minSize = Math.min(...impactSizes);
            
            const normalizedSizes = impactSizes.map(s => {{
                const normalized = (s - minSize) / (maxSize - minSize || 1);
                return 10 + normalized * 35;
            }});
            
            const hoverTexts = filteredData.map(d => 
                `${{d.City}}<br>State: ${{d.State}}<br>Magnitude: ${{d.Magnitude}}<br>Depth: ${{d.Depth_km}} km<br>Year: ${{d.Year}}`
            );
            
            const trace = {{
                lat: filteredData.map(d => d.Lat),
                lon: filteredData.map(d => d.Lon),
                mode: 'markers',
                type: 'scattermapbox',
                marker: {{
                    size: normalizedSizes,
                    color: filteredData.map(d => d.Magnitude),
                    colorscale: [[0, '#FFD700'], [0.3, '#FF8C00'], [0.7, '#FF4500'], [1, '#DC143C']],
                    showscale: true,
                    opacity: 0.95,
                    line: {{ color: '#FFF', width: 1.5 }},
                    colorbar: {{ title: '🔴 Magnitude', thickness: 20, len: 0.8, tickfont: {{ color: '#00d4ff' }} }}
                }},
                text: filteredData.map(d => d.City),
                hovertext: hoverTexts,
                hoverinfo: 'text',
                customdata: filteredData.map(d => d.City)
            }};
            
            const layout = {{
                title: {{ text: '<b>🎯 STRATEGIC SEISMIC TERMINAL 🎯</b>', x: 0.5, y: 0.97, font: {{ color: '#FF4500', size: 24, family: 'monospace' }} }},
                mapbox: {{ 
                    style: 'open-street-map',
                    center: {{ lat: 22.5, lon: 82.0 }},
                    zoom: 4.5,
                    uirevision: 'constant'
                }},
                paper_bgcolor: '#050505',
                plot_bgcolor: '#050505',
                font: {{ color: '#00d4ff', family: 'monospace' }},
                margin: {{ r: 0, t: 5, l: 0, b: 0 }},
                autosize: true,
                hovermode: 'closest',
                dragmode: 'zoom'
            }};
            
            return {{ data: [trace], layout }};
        }}
        
        function generateTrendGraph(cityName) {{
            const zoneName = getZoneName(cityName);
            const zoneData = earthquakeData.filter(d => d.City.startsWith(zoneName));
            
            const stats = {{}};
            for (let year = 2007; year <= 2017; year++) {{
                const yearData = zoneData.filter(d => d.Year === year);
                stats[year] = {{
                    count: yearData.length,
                    avg_mag: yearData.length > 0 
                        ? yearData.reduce((sum, d) => sum + d.Magnitude, 0) / yearData.length 
                        : 0
                }};
            }}
            
            const years = Object.keys(stats).map(Number).sort((a, b) => a - b);
            const counts = years.map(y => stats[y].count);
            const mags = years.map(y => stats[y].avg_mag);
            
            const trace1 = {{
                x: years, 
                y: counts,
                name: 'Frequency',
                type: 'scatter',
                mode: 'lines+markers',
                line: {{ color: '#00d4ff', width: 2, shape: 'spline' }},
                yaxis: 'y'
            }};
            
            const trace2 = {{
                x: years, 
                y: mags,
                name: 'Avg Magnitude',
                type: 'scatter',
                mode: 'lines+markers',
                line: {{ color: '#FF4500', width: 2, dash: 'dot', shape: 'spline' }},
                yaxis: 'y2'
            }};
            
            const layout = {{
                title: {{ text: `TREND: ${{zoneName}}`, font: {{ color: '#FF4500', size: 14 }} }},
                paper_bgcolor: '#050505',
                plot_bgcolor: '#050505',
                font: {{ color: '#00d4ff', family: 'monospace', size: 10 }},
                margin: {{ l: 35, r: 35, t: 35, b: 30 }},
                yaxis: {{ 
                    title: 'Freq',
                    titlefont: {{ color: '#00d4ff', size: 10 }},
                    tickfont: {{ color: '#00d4ff', size: 9 }}
                }},
                yaxis2: {{ 
                    title: 'Mag',
                    titlefont: {{ color: '#FF4500', size: 10 }},
                    tickfont: {{ color: '#FF4500', size: 9 }},
                    overlaying: 'y',
                    side: 'right'
                }},
                hovermode: 'x unified',
                autosize: true,
                showlegend: false
            }};
            
            return {{ data: [trace1, trace2], layout }};
        }}
        
        function updateDashboard() {{
            const minYear = parseInt(document.getElementById('year-min').value);
            const maxYear = parseInt(document.getElementById('year-max').value);
            
            if (minYear > maxYear) {{
                document.getElementById('year-min').value = maxYear;
            }}
            
            document.getElementById('year-display').textContent = minYear + ' - ' + maxYear;
            const mapFig = generateMap(minYear, maxYear);
            Plotly.newPlot('seismic-map', mapFig.data, mapFig.layout, {{ responsive: true }});
        }}
        
        document.addEventListener('DOMContentLoaded', function() {{
            console.log('🚀 Initializing dashboard...');
            
            const mapFig = generateMap(2007, 2017);
            Plotly.newPlot('seismic-map', mapFig.data, mapFig.layout, {{ responsive: true }});
            
            const trendFig = generateTrendGraph(earthquakeData[0].City);
            Plotly.newPlot('seismic-trend-graph', trendFig.data, trendFig.layout, {{ responsive: true }});
            
            displayGeologicalImage(earthquakeData[0].City);
            
            document.getElementById('year-min').addEventListener('input', updateDashboard);
            document.getElementById('year-max').addEventListener('input', updateDashboard);
            
            const mapElement = document.getElementById('seismic-map');
            mapElement.on('plotly_hover', function(data) {{
                if (data.points && data.points.length > 0) {{
                    const hoveredCity = data.points[0].customdata;
                    const trendFig = generateTrendGraph(hoveredCity);
                    Plotly.newPlot('seismic-trend-graph', trendFig.data, trendFig.layout, {{ responsive: true }});
                    displayGeologicalImage(hoveredCity);
                }}
            }});
            
            console.log('✅ Dashboard initialized');
        }});
    </script>
</body>
</html>"""
    
    if output_dir is None:
        output_dir = '.'
    
    if not os.path.exists(output_dir) and output_dir != '.':
        os.makedirs(output_dir)
    
    output_path = os.path.join(output_dir, filename)
    
    with open(output_path, 'w', encoding='utf-8') as f:
        f.write(html_content)
    
    file_size_kb = len(html_content) / 1024
    
    print(f"\n{{'='*75}}")
    print(f"✅ DASHBOARD EXPORTED SUCCESSFULLY")
    print(f"{{'='*75}}")
    print(f"📁 Location: {os.path.abspath(output_path)}")
    print(f"📊 Dataset: {len(df_100)} earthquake records")
    print(f"📦 File size: {file_size_kb:.1f} KB")
    print(f"{{'='*75}}\n")
    
    print("✨ LAYOUT: 50-50 Split with Trend & Image")
    print("   ✓ Left: War room alert map (Yellow-Orange-Red-Crimson)")
    print("   ✓ Right top: Trend graph (Frequency + Magnitude)")
    print("   ✓ Right bottom: 3D geological image")
    print("   ✓ Bottom: Year-range slider (2007-2017)")
    print("   ✓ Hover on map to update trend + image\n")
    
    return output_path

print("\n🚀 EXPORTING SEISMIC DASHBOARD...")
export_seismic_dashboard_html('seismic_dashboard_interactive.html')
print("✅ Dashboard ready with 50-50 layout!\n")
```
